```python
from builtins import bytes

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
df1 = pd.read_csv('robota_ua.csv')
df1["site"] = "robota.ua"
df1
```

```python
df2 = pd.read_csv('work_ua.csv')
df2.drop_duplicates()
df2["site"] = "work.ua"
df2

result_df = pd.concat([df1, df2], ignore_index=True)
result_df
```

```python
def calculate_relevance(row):
    score = 0
    for col in row.index:
        if col == 'info':
            if pd.notnull(row[col]) and len(str(row[col])) > 10:
                score += 1
        elif col == 'cities':
            if pd.notnull(row[col]) and len(str(row[col]).split(",")) > 1:
                score += 1
        elif pd.api.types.is_bool_dtype(result_df[col]):
            if row[col]:
                score += 1
        elif pd.notnull(row[col]) and row[col] != '':
            score += 1
    return score

result_df['relevance'] = result_df.apply(calculate_relevance, axis=1)
```

```python
result_df.sort_values(by='relevance', ascending=False, inplace=True)
result_df
```

```python
result_df.to_csv('result_df.csv')
```
